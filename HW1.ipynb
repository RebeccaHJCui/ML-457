## Part 6

```python
transformed_image  = np.dot(T,img_average_channel)
print(transformed_image)
plt.imshow(transformed_image)
# The original image is matrix with multiple values in rows and columns, and the image looks quiet clear.
# The dot product multiply each row (or column) by the row's (or column's) transpose and sum up all values.
# This gives us 1 row (or column) of values at the end.
# Since we have one row/column values here,  the image looks "mixed-together" with all its colors in each pixel.
```

## Part 7

2. Compute the inverse and store the matrix in `Tinv`.
3. Perform the inverse of the scrambling operation and plot the result to show that you retrieved the original image.

```python
#1 yes, matrix T is invertible, we can show the invertibilty by calculating its determiant. if its invertible,det(A) ≠ 0.
np.linalg.det(T)
print("Invert of T")
print(T)
# apply invert, should got the same matrix like the previous one
Tinv =np.linalg.inv(T)
print(Tinv)
plt.imshow(Tinv@transformed_image)
```
